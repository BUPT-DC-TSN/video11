```python
'''
将需要识别的视频和这个代码放在同一个文件夹下,文件名可以
在video_path变量中更改,interval变量控制每多少帧视频读
取一次数据,数据保存在result.txt中,仅有数字,不识别小数点
和负号等

使用截取视频的代码确定正确的截取区间后,替换掉box变量,可
以先用较大的interval测试是否可以正确

'''
import cv2#视频转图片
import pytesseract
import numpy
import os
from PIL import Image

name_of_result_file = 'result'#保存截图结果的图片文件名,可修改

result = ''#保存结果的字符串

#旧变量用于识别出错时替代错误结果,默认第一个识别结果不出错
text,text_old = '',''

video_path = '2.mp4'  # 视频地址
interval = 30 #每多少帧截一次数据
num_of_data = 0 #一共多少数据
n = 0

output_dir = "frames"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)


vid = cv2.VideoCapture(video_path)#打开视频

while vid.isOpened():

    is_read, frame = vid.read()
    if is_read :
        n = n+1
        if n == interval:
            n = 1
            num_of_data = num_of_data + 1 #数据个数记录
            image_org = Image.fromarray(cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)).convert('L')
#             image_org = image_org.transpose(Image.ROTATE_180)
            region = image_org.crop(box)
            frame_file = os.path.join(output_dir, f"frame_{num_of_data:04d}.jpg")
            region.save(frame_file)
            print(f"frame_{num_of_data:04d}.jpg")
    else:
        break
```

```python
import cv2
import os
import numpy as np

# 设置输入和输出文件夹路径
input_dir = "frames"
output_dir = "output_images"

# 创建输出文件夹（如果不存在）
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 遍历输入文件夹中的所有图像文件
for filename in os.listdir(input_dir):
    # 读取图像
    img_path = os.path.join(input_dir, filename)
    img = cv2.imread(img_path)
    
    # 将滤波后的图像转换为灰度图像
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 将灰度图像进行二值化
    ret, thresh_img = cv2.threshold(gray_img, 175, 255, cv2.THRESH_BINARY)
    
    # 对图像进行高斯滤波
    blur_img = cv2.GaussianBlur(thresh_img, (5, 5), 0)
    
    kernel = np.array([[0, -1, 0],[-1, 5, -1],[0, -1, 0]], np.float32)
    dst = cv2.filter2D(blur_img, -1, kernel=kernel)
    
    
    
    
    # 保存二值化后的图像
    output_path = os.path.join(output_dir, filename)
    cv2.imwrite(output_path, dst)
```

```python
import os
import cv2
import numpy as np
import pytesseract
import numpy
from PIL import Image

# 遍历文件夹中的所有图像文件
input_dir = "output_images"
for filename in os.listdir(input_dir):
    # 读取图像
    img_path = os.path.join(input_dir, filename)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    text = pytesseract.image_to_string(img)
    
    # 输出识别结果
    print(filename, "识别结果：", text)
```
