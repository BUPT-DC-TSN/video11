```python
from paddleocr import PaddleOCR, draw_ocr
import os
folder_path = "frames"
# Paddleocr目前支持的多语言语种可以通过修改lang参数进行切换
# 例如`ch`, `en`, `fr`, `german`, `korean`, `japan`
ocr = PaddleOCR(use_angle_cls=True, lang="en")  # need to run only once to download and load model into memory
for root, dirs, files in os.walk(folder_path):
    # 打印当前目录路径
    # 打印当前目录下的所有文件路径
    for file in files:
        file_path = os.path.join(root, file)
        print(file_path)
        result = ocr.ocr(file_path, cls=True)
        for idx in range(len(result)):
            res = result[idx]
            for line in res:
                print(line[1][0])
                result = line[1][0]
                with open("output.txt", "a",) as file:
                    file.write(result+'\n')

```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.lines as lines

# 创建画布和坐标轴
fig, ax = plt.subplots()

# 绘制路由器
router = patches.Rectangle((1, 1), 4, 2, linewidth=2, edgecolor='black', facecolor='white')
ax.add_patch(router)

# 绘制网卡1
nic1 = patches.Rectangle((0.5, 2.5), 1, 1, linewidth=2, edgecolor='black', facecolor='white')
ax.add_patch(nic1)

# 绘制网卡2
nic2 = patches.Rectangle((4.5, 2.5), 1, 1, linewidth=2, edgecolor='black', facecolor='white')
ax.add_patch(nic2)

# 绘制网线
cable = lines.Line2D((1.5, 4.5), (3, 3), linewidth=2, color='black')
ax.add_line(cable)

# 设置坐标轴范围和标题
ax.set_xlim(0, 6)
ax.set_ylim(0, 5)
ax.set_title('Router and NICs')

# 显示图形
plt.show()
```
